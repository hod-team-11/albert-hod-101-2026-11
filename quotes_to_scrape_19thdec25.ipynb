Bonus Exercice : build a "crawler" script that automatically detects and follows pagination links to scrape the entire website content.  Website : https://quotes.toscrape.com/

```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"
response = requests.get(url)

# Specify "html.parser" to avoid warnings and ensure consistent behavior
soup = BeautifulSoup(response.text, "html.parser")
```

* The quote text itself (e.g., "The world as we have created it is...").

```python
quotes = soup.find_all("span", class_="text")

for quote in quotes:
    print(quote.text)
```

* The name of the author (e.g., Albert Einstein)

```python
# Get all quotes first
quotes = soup.find_all("div", class_="quote")

for quote in quotes:
    # Extract the quote text
    text = quote.find("span", class_="text").text

    # Extract the author name
    author = quote.find("small", class_="author").text

    print(f"{author}")
```

* A list of associated tags (e.g., ['change', 'deep-thoughts', 'thinking', 'world']).

```python
all_tags = soup.find_all("a", class_="tag")
for tag in all_tags:
    print(tag.text)
```

Store the extracted data in a Python list of dictionaries


```python
tags_list = [tag.text for tag in soup.find_all("a", class_="tag")]
print(tags_list)
```
